# **Task 1: Data Cleaning & Preprocessing**

```python
# 1️⃣ Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
# 2️⃣ Upload Dataset (Colab)
from google.colab import files
uploaded = files.upload()

# Load dataset
df = pd.read_csv("Titanic-Dataset.csv")

print("Dataset Loaded Successfully ✅")
print("\nFirst 5 rows:")
display(df.head())
```

```python
# =========================================
# 3️⃣ Basic Information
# =========================================
print("\nDataset Info:")
print(df.info())

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# =========================================
# 4️⃣ Handle Missing Values
# =========================================

# Fill Age with Median
df['Age'].fillna(df['Age'].median(), inplace=True)

# Fill Embarked with Mode
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Drop Cabin (Too many null values)
df.drop(columns=['Cabin'], inplace=True)

print("\nMissing Values After Treatment:")
print(df.isnull().sum())
```

```python
# =========================================
# 5️⃣ Encode Categorical Features
# =========================================

# Label Encoding for Sex
le = LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex'])

# One-Hot Encoding for Embarked
df = pd.get_dummies(df, columns=['Embarked'], drop_first=True)

print("\nAfter Encoding:")
display(df.head())
```

```python
# =========================================
# 6️⃣ Outlier Detection & Removal
# =========================================

plt.figure(figsize=(10,5))
sns.boxplot(data=df[['Age','Fare']])
plt.title("Boxplot for Age and Fare")
plt.show()

# Removing Outliers using IQR method
def remove_outliers(column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[column] >= lower) & (df[column] <= upper)]

df = remove_outliers('Fare')
df = remove_outliers('Age')

print("Outliers Removed ✅")
print("New Shape:", df.shape)
```

# **Survival Count Plot**

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.show()
```

# **Survival by Gender**

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival by Gender")
plt.show()
```

# **Age Distribution**

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title("Age Distribution")
plt.show()
```

# **Correlation Heatmap**

```python
plt.figure(figsize=(10,6))
# Drop non-numeric and identifier columns before calculating correlation
df_numeric = df.drop(columns=['Name', 'Ticket', 'PassengerId'], errors='ignore')
sns.heatmap(df_numeric.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# =========================================
# 7️⃣ Feature Scaling (Standardization)
# =========================================

scaler = StandardScaler()
numerical_cols = ['Age', 'Fare', 'SibSp', 'Parch']

df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

print("\nAfter Scaling:")
display(df.head())
```

```python
# =========================================
# 8️⃣ Final Clean Dataset
# =========================================

df.to_csv("Cleaned_Titanic.csv", index=False)
print("\nCleaned Dataset Saved Successfully 🚀")
```
